# Image Data Augmentation Using Deep Convolutional Generative Adversarial Network (DCGAN)
# Model - Check Training Results

## Load Libraries

```python
%matplotlib inline

import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## While the GAN training is in progress, sample images and masks for saved to file at regular intervals after a fixed
## number of iterations. Show these images to check the training progress.

```python
# Load the image files. These were saved as numpy arrays with 5 channels - image RGBA (4 channels) + mask (1 channel).
image_list = []

# Loop through all the directories
for root, dirs, files in os.walk('data/gan_images'):
    for afile in files:
        a_image = np.load(os.path.join(root, afile))
        image_list.append([afile, a_image])
gan_images = image_list
num_files = len(gan_images)
num_files
```

## Show a sample of generated and training image and mask.

```python
# Plot a generated image and mask. Also show the real image and mask used in the GAN training.
plt.figure(figsize=(10,10))

a_sample = gan_images[0]
a_img = a_sample[1][:, :, 0:3]
a_mask = a_sample[1][:, :, 4]

plt.subplot(221)
plt.imshow(a_img.astype(np.uint))
plt.title(a_sample[0])
plt.subplot(223)
plt.imshow(a_mask.astype(np.uint))

a_sample = gan_images[4]
a_img = a_sample[1][:, :, 0:3]
a_mask = a_sample[1][:, :, 4]

plt.subplot(222)
plt.imshow(a_img.astype(np.uint))
plt.title(a_sample[0])
plt.subplot(224)
plt.imshow(a_mask.astype(np.uint))
plt.show()
```

## Show the generated images and mask over training iterations.

```python
all_list = []
for i in range(len(gan_images)):
    a_sample = gan_images[i]
    a_img = a_sample[1][:, :, 0:3]
    all_list.append([a_sample[0], a_img])
all_list.sort(key=lambda x:x[0])
all_list = all_list[:5] 

mask_list = []
for i in range(len(gan_images)):
    a_sample = gan_images[i]
    a_mask = a_sample[1][:, :, 4]
    mask_list.append([a_sample[0], a_mask])
mask_list.sort(key=lambda x:x[0])
mask_list = mask_list[:5]
all_list = all_list + mask_list

fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(20, 7))
i = 0
for row in ax:
    for col in row:
        if i < 5 :
            col.imshow(all_list[i][1].astype(np.uint), aspect='auto', cmap='viridis')
        else:
            col.imshow(all_list[i][1].astype(np.uint), aspect='auto', cmap='gray')
        col.set_title(all_list[i][0])
        i += 1

plt.show()
```
